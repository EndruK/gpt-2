```python
import os
import random
from shutil import copyfile
```

```python
path = "/mnt/media/Corpora/extern/Github_Java_Corpus/java_projects"
```

```python
file_list = []
cnt = 0
for root, dirs, files in os.walk(path):
    for file in files:
        if cnt % 500 == 0:
            print(str(cnt) + " files done")
        fq_file_path = os.path.join(root, file)
        file_list.append(fq_file_path)
        cnt += 1
```

```python
sample_size = 50000
output_path = "/mnt/media/Corpora/extern/Github_Java_Corpus/subset"
```

```python
subset = [file_list[i] for i in sorted(random.sample(range(len(file_list)), sample_size))]
```

```python
for i, file_path in enumerate(subset):
    if i % 200 == 0:
        print(str(i) + " / " + str(len(subset)))
    rel_file_path = file_path.split(path+"/")[1]
    
    target_file_path = os.path.join(output_path, rel_file_path)
    target_folder_path = "/".join(target_file_path.split("/")[:-1])
    if not os.path.exists(target_folder_path):
        os.makedirs(target_folder_path)
    copyfile(file_path, target_file_path)
```
